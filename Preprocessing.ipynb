# Train Preprocess

```python
train_path = '../train.tsv' # Здесь исходная тренировочная выборка
test_path = '../test.tsv' # исходная тестовая
train_preprocessed_path = './Roma/train2.csv' # Сюда сохраняем предобработанную
test_preprocessed_path = '/home/jupyter-admin/Roma/test2.csv'
```

```python
data = pd.read_csv(train_path, sep='\t', names=['phrase', 'context', 'target'])
```

```python
data['context'] = data['context'].apply(ast.literal_eval)
```

```python
series = []
for row in data['context']:
    text = ' '
    for d in row:
        text += ('user: ' + d['user'] + '. ' if d['user'] else '') + ('marusia: ' + d['marusia'] + '. ' if d['marusia'] else '')
    series.append(text)
data['context_text'] = series 
#data = data.drop('context', axis='columns')
data
```

```python
data.to_csv(train_preprocessed_path, index=False)
```

# Test preprocess

```python
test = pd.read_csv(test_path, sep='\t')
test.columns = ['phrase', 'context', 'id']
```

```python
test['context'] = test['context'].apply(ast.literal_eval)
```

```python
series = []
for row in test['context']:
    text = ' '
    for d in row:
        text += ('user: ' + d['user'] + '. ' if d['user'] else '') + ('marusia: ' + d['marusia'] + '. ' if d['marusia'] else '')
    series.append(text)
test['context_text'] = series 
test = test.drop('context', axis='columns')
test
```

```python
test.to_csv(test_preprocessed_path, index=False)
```
